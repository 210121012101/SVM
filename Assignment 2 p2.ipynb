```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split
from numpy import linalg as la
import seaborn as sns
import pylab
import scipy.stats as stats
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
# from svmutil import *
```

```python
df_train = pd.read_csv('train_new.csv',index_col='Sample ID')
df_test = pd.read_csv('test_new.csv',index_col='Sample ID')
```

```python
df_train.head()
```

```python
X = df_train.drop(columns='Class',axis=1,)
y = df_train['Class']
```

## Course grid search

```python
param_grid1 = {'C': [0.1,1,10,100], 
              'gamma': [0.01,0.1,1,10],
              'kernel': ['rbf','linear','poly','sigmoid'],
             'degree':[1,2,3,4,5,6]}
```

```python
clf_course = GridSearchCV(SVC(),param_grid=param_grid1, cv=5)
```

```python
clf_course.fit(X,y)
```

```python
clf_course.best_params_
```

```python
clf_course.cv_results_
```

## fine grid search

```python
param_grid = {'C': [5,5.4,6.4,7,7.4], 
              'gamma': [0.02,0.03,0.04,0.06,0.07,0.08],
              'kernel': ['rbf']} 
```

```python
clf = GridSearchCV(SVC(C=5.4,gamma=0.04),param_grid=param_grid, cv=15)
```

```python
clf.fit(X,y)
```

```python
clf.best_params_
```

```python
clf.cv_results_
```

## using model on test data

```python
df_test.head()
```

```python
X_t = np.array(df_test)
```

```python
y = clf.predict(X_t)
```

```python
file = pd.DataFrame({'Sample ID':range(1,len(df_test)+1),'Class':y})
file.to_csv('a2_kaggle1.csv')
```
